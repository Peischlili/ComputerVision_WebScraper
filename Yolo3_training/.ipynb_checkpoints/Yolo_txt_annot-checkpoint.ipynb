```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import random
```

```python
# Defining global variable path that contains train / train_resized folders
img_folder = "/Users/peisch/code/WebScraper/Achive_images/screenshots"

# Function to load folder into arrays and then it returns that same array
def loadAnnotation(path):
    txt_files = sorted([os.path.join(path, "LabelsYoloformat", file)
         for file in os.listdir(path + "/LabelsYoloformat") if file.endswith('.txt')])
    return txt_files
```

```python
files = loadAnnotation(img_folder)
```

```python
def rewrite_txt(filelist):
    
    for each_file in filelist:
        os.chdir(f"{img_folder}/LabelsYoloformat/")
        filename = each_file.replace(f"{img_folder}/LabelsYoloformat/", "")
        f = open(filename,"r")
        content = f.read()
        f.close()
        
        os.chdir(img_folder)
        try:
            os.makedirs('YoloLabel')
        except FileExistsError:
            print("Directory 'YoloLabel' already exists. Annotation processing ... ")
         
        os.chdir(f"{img_folder}/YoloLabel/")
        # create new txt file
        new_file = open(filename, "w")
        new_content = content.replace("Dress", "0")
        n = new_file.write(new_content)
        new_file.close()
        print(new_content)
```

```python
rewrite_txt(files)
```

```python
# create test configuration file
def create_test_txt(filelist):
    test_list = []
    for each_file in filelist:
        os.chdir(f"{img_folder}/LabelsYoloformat/")
        filename = each_file.rstrip('.txt').replace(f"{img_folder}/LabelsYoloformat/", "")
        new_name = f"data/obj/{filename}.jpg"
        test_list.append(new_name)
    # convert list to string for txt.file content
    os.chdir(img_folder)
    test_str = '\n'.join(test_list)
    file = open("test.txt", "w")
    n = file.write(test_str)
    file.close()
    return test_list
```

```python
create_test_txt(files)
```
